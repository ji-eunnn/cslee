```python
from keras import models
from keras import layers
import numpy as np 
```

<br>
<br>

# AND 연산

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[0],[0],[1]])

model = models.Sequential()
model.add(layers.Dense(1, activation='sigmoid', input_shape=(2,)))

model.compile(optimizer='rmsprop', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

model.fit(x, y, epochs=3000, batch_size=1)
```

```python
print(model.predict(np.array([[0,0]])).round())
print(model.predict(np.array([[0,1]])).round())
print(model.predict(np.array([[1,0]])).round())
print(model.predict(np.array([[1,1]])).round())
```

AND 연산은 단층신경망으로 학습이 정상적으로 수행됩니다.

<br>
<br>

# XOR 연산

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

model = models.Sequential()
model.add(layers.Dense(1, activation='sigmoid', input_shape=(2,)))

model.compile(optimizer='rmsprop', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

model.fit(x, y, epochs=3000, batch_size=1)
```

XOR 연산을 단층신경망으로 하면 학습이 불가능합니다.
<br>
<br>
<br>

```python
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(2,))) # 히든 레이어 추가
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

model.fit(x, y, epochs=1000, batch_size=1)
```

```python
print(model.predict(np.array([[0,0]])).round())
print(model.predict(np.array([[0,1]])).round())
print(model.predict(np.array([[1,0]])).round())
print(model.predict(np.array([[1,1]])).round())
```

단층신경망에 16개 노드를 가진 히든 레이어를 추가하여 다층신경망으로 변경하였습니다. 보통 노드의 개수는 2의 배수로 해야 행렬 계산에 유리합니다. 이제 XOR 문제도 정상적으로 학습이 가능해졌습니다.
